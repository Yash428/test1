```python
import numpy as np
import matplotlib.pyplot as plt 
import keras as ke
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import pandas as pd
from keras.layers import *  
from sklearn.model_selection import train_test_split
```

```python
#simple wire
df = pd.DataFrame({'x1':[1,1,1,0,0,0],'y':[1,1,1,0,0,0]})
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('y',axis=1)
y = df['y']

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
model = Sequential()
input = Input(shape=(1,))
model.add(input)
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=2)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
from sklearn.datasets import make_blobs, make_moons, make_circles
X1, y1 = make_blobs(n_samples=200, centers=2,random_state=0, cluster_std=0.60)
y1 = np.where(y1 <= 0, -1, 1)
X3, y3 = noisy_circles = make_circles(n_samples=200, factor=.5, noise=.05)
y3 = np.where(y3 <= 0, -1, 1)
plt.scatter(X3[:, 0], X3[:, 1], c=y3, s=50, cmap='winter', alpha=.5)
plt.title("Dataset 3")
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min()-0.1, X[:, 0].max()+0.1
    y_min, y_max = X[:, 1].min()-0.1, X[:, 1].max()+0.1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))

    x_in = np.c_[(xx.ravel(), yy.ravel())]

    y_pred = model.predict(x_in)

    if len(y_pred[0]) > 1:
        print('Doing multi-class')
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print('Doing binary classification')
        y_pred = np.round(y_pred).reshape(xx.shape)

    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```
